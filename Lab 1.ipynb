## 2.1.1: Plate Scale (Ronchi Mask Resolution)

#### Determine an optimal imaging system with a plate scale that would allow to resolve two successive lines of the provided Ronchi mask. 


### Lens Setup:
- **Camera → Lens #2 (75 mm) → Lens #3 (100 mm) → Ronchi Mask**

### Explanation of Lens Placement:
- **Distance between lenses**: The lenses should be placed as close as possible to each other to minimize the effect of light divergence or convergence.
- **Distance from the camera**: The **75 mm lens** should be placed about **1-2 focal lengths** away from the camera. The object (Ronchi mask) should be placed far away on the optical breadboard to simulate an object at infinity.

### Magnification:
For an object at infinity, the magnification \(M\) is calculated as the ratio of the focal lengths:
$$
M = \frac{f_2}{f_1}
$$
Where:
- $f_1 = 75 \, \text{mm}$ (first lens)
- $f_2 = 100 \, \text{mm}$ (second lens)


### Effective Focal Length (EFL):
The combined **effective focal length** (EFL) of the two lenses is calculated using the thin lens equation:
$$
\frac{1}{\text{EFL}} = \frac{1}{f_1} + \frac{1}{f_2}
$$

### Plate Scale:
The plate scale is calculated as:
$$
\text{Plate Scale (arcseconds/pixel)} = \frac{206,265}{\text{EFL}} \times \text{Pixel Size (mm)}
$$
Where:
- Pixel size of the Blackfly camera = 2.4 µm = 0.0024 mm

#### Summary for 2.1.1:

- Effective Focal Length (EFL): 42.86 mm
- Magnification: 1.33x
- Plate Scale: 11.55 arcseconds/pixel
- Distance from Camera to Lens #2 (75 mm): 75 mm
- Distance between Lens #2 and Lens #3: 10 mm
- Ronchi Mask: Far from the system (acting as an object at infinity).

## 2.1.2: Field of View (Imaging Jupiter)
#### Determine an optimal imaging system that is capable of imaging the entire extent of Jupiter in a single exposure.


### Lens Setup:
- **Camera → Lens #1 (50 mm) → Lens #2 (75 mm) → Jupiter Image**

### Explanation of Lens Placement:
- **Distance between lenses**: Place the **lenses as close as possible** to each other. Leave about **1-2 focal lengths** of the **50 mm lens** between the camera and the first lens.
- **Distance from the camera**: The object (Jupiter image card) should be placed far away on the optical breadboard to simulate an object at infinity.

### Magnification:
The magnification \(M\) for this system is:
$$
M = \frac{f_2}{f_1}
$$

### Effective Focal Length (EFL):
The effective focal length is calculated using the thin lens equation:
$$
\frac{1}{\text{EFL}} = \frac{1}{f_1} + \frac{1}{f_2}
$$

### Field of View (FOV):
The field of view is calculated as:
$$
\text{FOV (degrees)} = 2 \times \arctan\left(\frac{\text{Sensor Dimension (mm)}}{2 \times \text{EFL (mm)}}\right)
$$
Where:
- Sensor width of the Blackfly camera = 7.37 m

#### Calculation:
$$
\theta_{\text{diffraction}} = \frac{1.22 \times 550 \times 10^{-9}}{25.4 \times 10^{-3}} = 2.64 \times 10^{-5} \, \text{radians}
$$
Converting to arcseconds:
$$
\theta_{\text{diffraction}} = 2.64 \times 10^{-5} \times 206,265 = 5.45 \, \text{arcseconds}
$$

#### Comparison:
- **Diffraction limit**: 5.45 arcseconds.
- **Plate scale resolution**: 11.55 arcseconds/pixel.

The system’s resolution is **dominated by the plate scale**, and is not diffraction-limited. However, the system still provides good resolution for the tasks at hand (resolving Jupiter's red spot and the globular cluster simulation).

```python
import math

# Constants
arcseconds_per_radian = 206265  # Conversion factor
pixel_size = 0.0024  # Pixel size of the camera in mm
sensor_width = 7.37  # Sensor width in mm

# 2.1.1: Plate Scale (Ronchi Mask Resolution)
f1_211 = 75  # Focal length of the first lens in mm
f2_211 = 100  # Focal length of the second lens in mm

# Effective Focal Length (EFL)
efl_211 = 1 / (1 / f1_211 + 1 / f2_211)
print(f"Effective Focal Length (EFL) for 2.1.1: {efl_211:.2f} mm")

# Plate Scale
plate_scale_211 = (arcseconds_per_radian / efl_211) * pixel_size
print(f"Plate Scale for 2.1.1: {plate_scale_211:.2f} arcseconds/pixel")

# Magnification
magnification_211 = f2_211 / f1_211
print(f"Magnification for 2.1.1: {magnification_211:.2f}")

# 2.1.2: Field of View (Imaging Jupiter)
f1_212 = 50  # Focal length of the first lens in mm
f2_212 = 75  # Focal length of the second lens in mm

# Effective Focal Length (EFL)
efl_212 = 1 / (1 / f1_212 + 1 / f2_212)
print(f"Effective Focal Length (EFL) for 2.1.2: {efl_212:.2f} mm")

# Field of View (FOV)
fov_212 = 2 * math.degrees(math.atan(sensor_width / (2 * efl_212)))
print(f"Field of View (FOV) for 2.1.2: {fov_212:.2f} degrees")

# Magnification
magnification_212 = f2_212 / f1_212
print(f"Magnification for 2.1.2: {magnification_212:.2f}")

# 2.1.3: Field of View & Plate Scale (Resolving Jupiter's Red Spot)
f1_213 = 75  # Focal length of the first lens in mm
f2_213 = 100  # Focal length of the second lens in mm

# Effective Focal Length (EFL)
efl_213 = 1 / (1 / f1_213 + 1 / f2_213)
print(f"Effective Focal Length (EFL) for 2.1.3: {efl_213:.2f} mm")

# Plate Scale
plate_scale_213 = (arcseconds_per_radian / efl_213) * pixel_size
print(f"Plate Scale for 2.1.3: {plate_scale_213:.2f} arcseconds/pixel")

# Field of View (FOV)
fov_213 = 2 * math.degrees(math.atan(sensor_width / (2 * efl_213)))
print(f"Field of View (FOV) for 2.1.3: {fov_213:.2f} degrees")

# Magnification
magnification_213 = f2_213 / f1_213
print(f"Magnification for 2.1.3: {magnification_213:.2f}")
```
