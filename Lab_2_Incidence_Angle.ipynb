```python
import numpy as np
import pandas as pd

# Known Neon Values & Experimental Neon Values
data = {
    'X Values': [588.375, 586.750, 590.750, 591.000, 591.125, 593.000, 587.500],
    'Wavelength (nm)': [600.0365, 602.7263, 615.6283, 618.6716, 619.1023, 626.8232, 629.7388] }

# Groove spacing
d = 1.6667e-6

# Diffraction Magnitude
m = 1

# Pixel to Radians Conversion
proportionality_factor = 2e-5

# Calculate diffraction of theta_d
theta_d = np.array(data['X Values']) * proportionality_factor

# Convert to meters
wavelengths_m = np.array(data['Wavelength (nm)']) * 1e-9

# Useing Grating Equation to find Angle
theta_i = np.arcsin((m * wavelengths_m / d) - np.sin(theta_d))

# Convert the incident angle from radians to degrees
theta_i_degrees = np.degrees(theta_i)

# Dataframe
result_df = pd.DataFrame({
    'X Values': data['X Values'],
    'Wavelength (nm)': data['Wavelength (nm)'],
    'Theta_d (radians)': theta_d,
    'Theta_i (incident angle, radians)': theta_i,
    'Theta_i (incident angle, degrees)': theta_i_degrees
})

result_df

```

```python
# Calculate the average of the last column (incident angle in degrees)
average_theta_i_degrees = result_df.iloc[:, -1].mean()

print(f"Average of incident angles (degrees): {average_theta_i_degrees}")
```
