## 2.1.1: Plate Scale (Ronchi Mask Resolution)

#### Determine an optimal imaging system with a plate scale that would allow to resolve two successive lines of the provided Ronchi mask. 


### Lens Setup 1:
- **Camera → Lens #2 (75 mm) → Lens #3 (100 mm) → Ronchi Mask**

### Explanation of Lens Placement:
- **Distance between lenses**: The lenses should be placed as close as possible to each other to minimize the effect of light divergence or convergence.
- **Distance from the camera**: The **75 mm lens** should be placed about **1-2 focal lengths** away from the camera. The object (Ronchi mask) should be placed far away on the optical breadboard to simulate an object at infinity.

### Magnification:
For an object at infinity, the magnification \(M\) is calculated as the ratio of the focal lengths:
$$
M = \frac{f_2}{f_1}
$$
Where:
- $f_1 = 75 \, \text{mm}$ (first lens)
- $f_2 = 100 \, \text{mm}$ (second lens)


### Effective Focal Length (EFL):
The combined **effective focal length** (EFL) of the two lenses is calculated using the thin lens equation:
$$
\frac{1}{\text{EFL}} = \frac{1}{f_1} + \frac{1}{f_2}
$$

### Plate Scale:
The plate scale is calculated as:
$$
\text{Plate Scale (arcseconds/pixel)} = \frac{206,265}{\text{EFL}} \times \text{Pixel Size (mm)}
$$
Where:
- Pixel size of the Blackfly camera = 2.4 µm = 0.0024 mm

#### Summary for 2.1.1 (Setup 1):

- Effective Focal Length (EFL): 42.86 mm
- Magnification: 1.33x
- Plate Scale: 11.55 arcseconds/pixel
- Distance from Camera to Lens #2 (75 mm): 75 mm
- Distance between Lens #2 and Lens #3: 10 mm
- Ronchi Mask: Far from the system (acting as an object at infinity).

### Calculation of Plate Scale for Resolving Jupiter's Red Spot

We want to calculate the plate scale required to resolve the **5 mm red spot** on the **paper Jupiter**.

### Step 1: Calculate the Angular Size of the Red Spot

The angular size: 

$$\theta_{\text{arcseconds}} $$ 

of an object is given by the formula:


$$
\theta_{\text{arcseconds}} = \frac{\text{Physical Size of Object (mm)}}{\text{Distance to Object (mm)}} \times 206,265 = \frac{5mm}{1000mm} \times 206,265 = \text{1031.33 arcseconds}
$$

Where:
- The **physical size** of the red spot is 5 mm.
- The **distance to the object** (paper Jupiter) is 1,000 mm (1 meter).

### Step 2: Calculate the Required Plate Scale

The required plate scale to resolve the red spot over at least 2 pixels is:

$$
\text{Plate Scale (arcseconds/pixel)} = \frac{\theta_{\text{arcseconds}}}{\text{Number of Pixels}} = \frac{1031.33}{2} = \text{515.67}\frac{arcseconds}{pixels}
$$

We will assume we want the red spot to cover at least 2 pixels.

### Step 3: Compare to the Current Plate Scale

The current plate scale for our optical setup (75 mm + 100 mm lenses) is calculated as:

$$
\text{Current Plate Scale} = \frac{206,265}{\text{EFL}} \times \text{Pixel Size (mm)} = 11.55 \frac{arcseconds}{pixel}
$$

Where the effective focal length (EFL) is **42.86 mm** (from the previous calculation), and the pixel size is **2.4 µm** or **0.0024 mm**. Since our plate scale is finer we are able to clearly see the Red Spot on Jupiter.


```python
import math

# Constants
arcseconds_per_radian = 206265  # Conversion factor
pixel_size = 0.0024  # Pixel size of the camera in mm
sensor_width = 7.37  # Sensor width in mm

# 2.1.1: Plate Scale (Ronchi Mask Resolution)
f1_211 = 75  # Focal length of the first lens in mm
f2_211 = 100  # Focal length of the second lens in mm

# Effective Focal Length (EFL)
efl_211 = 1 / (1 / f1_211 + 1 / f2_211)
print(f"Effective Focal Length (EFL) for 2.1.1: {efl_211:.2f} mm")

# Plate Scale
plate_scale_211 = (arcseconds_per_radian / efl_211) * pixel_size
print(f"Plate Scale for 2.1.1: {plate_scale_211:.2f} arcseconds/pixel")

# Magnification
magnification_211 = f2_211 / f1_211
print(f"Magnification for 2.1.1: {magnification_211:.2f}")

# 2.1.2: Field of View (Imaging Jupiter)
f1_212 = 50  # Focal length of the first lens in mm
f2_212 = 75  # Focal length of the second lens in mm

# Effective Focal Length (EFL)
efl_212 = 1 / (1 / f1_212 + 1 / f2_212)
print(f"Effective Focal Length (EFL) for 2.1.2: {efl_212:.2f} mm")

# Field of View (FOV)
fov_212 = 2 * math.degrees(math.atan(sensor_width / (2 * efl_212)))
print(f"Field of View (FOV) for 2.1.2: {fov_212:.2f} degrees")

# Magnification
magnification_212 = f2_212 / f1_212
print(f"Magnification for 2.1.2: {magnification_212:.2f}")

# 2.1.3: Field of View & Plate Scale (Resolving Jupiter's Red Spot)
f1_213 = 75  # Focal length of the first lens in mm
f2_213 = 100  # Focal length of the second lens in mm

# Effective Focal Length (EFL)
efl_213 = 1 / (1 / f1_213 + 1 / f2_213)
print(f"Effective Focal Length (EFL) for 2.1.3: {efl_213:.2f} mm")

# Plate Scale
plate_scale_213 = (arcseconds_per_radian / efl_213) * pixel_size
print(f"Plate Scale for 2.1.3: {plate_scale_213:.2f} arcseconds/pixel")

# Field of View (FOV)
fov_213 = 2 * math.degrees(math.atan(sensor_width / (2 * efl_213)))
print(f"Field of View (FOV) for 2.1.3: {fov_213:.2f} degrees")

# Magnification
magnification_213 = f2_213 / f1_213
print(f"Magnification for 2.1.3: {magnification_213:.2f}")
```

## 2.2.1 Bias and Dark Current

### Objective: 
- Bias Level: Measure the bias level of the detector by collecting dark frames with short integration times. 
- Dark Current: Measure the dark current by collecting dark frames at varying exposure times.

```python
import numpy as np
from astropy.io import fits
import struct
import numpy as np
from glob import glob

import matplotlib.pyplot as plt

%matplotlib inline
def raw_to_fits(raw_file, shape=(1080, 1440), destination=None, ADC_bits=12, simu_artifacts=False):
    """
    Converts a .raw file to a FITS formatted image.
    raw_file : The file path of the raw image. Must have format of
        <prefix>_<texp>us.raw
    shape : A list or tuple with (# of rows, # of columns) for the image
    destination : an alternate destination/filename for the FITS image
       (default is to replace the .raw extension and save in the same folder)
    """
    try:
        # Extract timestamp or set a default value
        parts = raw_file.split("us.raw")[0].split("_")
        if parts and parts[-1].replace('.', '', 1).isdigit():
            tint = float(parts[-1]) / 1e6
        else:
            tint = 0  # Default value
    except (ValueError, IndexError) as e:
        print(f"Error extracting timestamp from filename '{raw_file}': {e}")
        tint = 0  # Default value or handle as appropriate

    # Get shape of image
    nrows, ncols = shape
    # Open raw file
    with open(raw_file, 'rb') as raw:
        raw_img = raw.read()
    # Check file size / format
    filesize = len(raw_img)
    if filesize == nrows * ncols:  # 8-bit
        size = 'B'
        file_bits = 8
    elif filesize == nrows * ncols * 2:  # 16-bit
        size = 'H'
        file_bits = 16
    else:  # No size match
        raise ValueError("The image shape provided does not match the length of the file")
    # Format binary string
    fmt = f"<{nrows * ncols}{size}"
    # Convert byte string to numpy array
    byte_array = np.array(struct.unpack(fmt, raw_img)).reshape(nrows, ncols)
    if file_bits > ADC_bits:
        byte_array = byte_array / 2 ** (file_bits - ADC_bits)

    if simu_artifacts:
        ny, nx = byte_array.shape

        # Adding dark
        dark_current = 1.25  # ADU/s
        byte_array += dark_current * tint

        # multiplying flat
        x_grid, y_grid = np.meshgrid(np.arange(nx), np.arange(ny))
        flat_field = np.exp(-0.5 * ((x_grid - nx // 2) ** 2 / (3 * nx) ** 2 + (y_grid - ny // 2) ** 2 / (3 * ny) ** 2))
        byte_array *= flat_field

        state = np.random.get_state()
        np.random.seed(0)

        # adding hot pixels
        hot_pix_indices = np.random.randint(0, high=nx * ny - 1, size=int(0.01 * nx * ny))
        hot_pix_indices = np.unravel_index(hot_pix_indices, (ny, nx))
        byte_array[hot_pix_indices] *= tint * np.clip(30 * np.random.randn(int(0.01 * nx * ny)), dark_current, 2 ** ADC_bits)

        # adding cold pixels
        cold_pix_indices = np.random.randint(0, high=nx * ny - 1, size=int(0.01 * nx * ny))
        cold_pix_indices = np.unravel_index(cold_pix_indices, (ny, nx))
        byte_array[cold_pix_indices] *= np.random.uniform(0, 0.5, size=int(0.01 * nx * ny))

        np.random.set_state(state)

        # Adding bias
        byte_array += 64

        # adding readnoise
        readnoise = 3  # ADU
        byte_array += readnoise * np.random.randn(ny, nx)

        byte_array = np.clip(byte_array, 0, 2 ** ADC_bits)

    # FITS file name
    if destination is None:
        fits_file = raw_file.replace('.raw', '.fits')
    else:
        fits_file = destination
    if simu_artifacts:
        fits_file = fits_file.replace(".fits", "_simu.fits")
    # Save as FITS
    hdu = fits.PrimaryHDU(byte_array)
    hdu.writeto(fits_file, overwrite=True)
```
